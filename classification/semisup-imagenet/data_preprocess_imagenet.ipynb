```python
import os
import sys
import _pickle as cPickle
import csv
import numpy as np
from PIL import Image
from tqdm import tqdm

from IPython.core.debugger import Tracer
```

```python
data_dir = '/tanData/datasets/imagenet/data'
output_dir = '/tanData/datasets/imagenet/data'
output_list_sup_path = 'train_sup.lst'
output_list_unsup_path = 'train_unsup.lst'
seed_val = 0
num_train_sup = 128000
```

```python
# Preparing the data step 1
with open(os.path.join(data_dir, 'train.lst'), 'r') as f:
    lines = f.readlines()
    tokens = [i.rstrip().split('\t') for i in lines]
    idx_label = dict((int(idx), label) for idx, label, _ in tokens)
    idx_name = dict((int(idx), name) for idx, _, name in tokens)
         
labels = set(idx_label.values())

num_train = 0
for root, dirs, files in os.walk(os.path.join(data_dir, 'train')):
    num_train += len(files)

num_train_sup_per_label = num_train_sup // len(labels)

ratio_unsup_sup = (num_train - num_train_sup) // num_train_sup

# select labeled data
data_rng = np.random.RandomState(seed_val)
indx = data_rng.permutation(range(1, num_train + 1))
indx_sup = []

with open(os.path.join(os.path.join(data_dir, output_list_sup_path)), "w") as list_writer_sup:
    fo_sup = csv.writer(list_writer_sup, lineterminator='\n', delimiter='\t')
    for c in tqdm(labels):
        c_count = 0
        for i in indx:
            if idx_label[i] == c and c_count < num_train_sup_per_label:
                row = [i, idx_label[i], idx_name[i]]
                for j in range(ratio_unsup_sup):
                    fo_sup.writerow(row)
                indx_sup.append(i)
                c_count += 1
            if c_count >= num_train_sup_per_label:
                break

count_val = 0
with open(os.path.join(os.path.join(data_dir, output_list_unsup_path)), "w") as list_writer_unsup:
    fo_unsup = csv.writer(list_writer_unsup, lineterminator='\n', delimiter='\t')
    for i in tqdm(indx):
        if i not in indx_sup:
            row = [i, idx_label[i], idx_name[i]]
            fo_unsup.writerow(row)
            count_val += 1

print(count_val)
```
